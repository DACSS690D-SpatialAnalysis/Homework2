```python
# Get gas stations in Lowell
import osmnx as ox

stations = ox.features_from_place("Lowell, Massachusetts, USA",
                                tags={'amenity': 'fuel'})
```

```python
stations.info()
stations.crs
```

```python
stations.reset_index(drop=False,inplace=True)
stations.head()
```

```python
colsToKeep=['element','name','geometry']
fuelLowell=stations[colsToKeep].copy()
```

```python
lowell_crs = "EPSG:26986"

fuelLowell = fuelLowell.to_crs(lowell_crs)
fuelLowell.info()
```

```python
fuelLowell.fillna({'name':'Unknown'},inplace=True)
fuelLowell['newname']=fuelLowell.index.astype(str)+"_"+fuelLowell.name
fuelLowell.set_index('newname',inplace=True)
fuelLowell.drop(columns=['name'],inplace=True)
fuelLowell.head()
```

```python
# Get schools in Lowell
schools = ox.features_from_place("Lowell, Massachusetts, USA",
                                tags={'amenity': 'school'})
schools.info()
```

```python
schools.reset_index(inplace=True)
schoolsLowell=schools[['element','name','geometry']].copy()
schoolsLowell.fillna({'name':'Unknown'},inplace=True)
schoolsLowell['newname']=schoolsLowell.index.astype(str)+"_"+schoolsLowell.name
schoolsLowell.set_index('newname',inplace=True)
schoolsLowell.drop(columns=['name'],inplace=True)
schoolsLowell = schoolsLowell.to_crs(lowell_crs)
schoolsLowell.head()
```

```python
base=lowell.explore(tiles='cartodbpositron',color='lightblue')
schoolsLowell.explore(m=base,color='k')
fuelLowell.explore(m=base,color='red')
```

Find gas stations that should be removed

```python
# minimal distances between gas stations and schools sorted lowest to highest
D_Matrix_fuel_school.min(axis=1).sort_values().head(10)
```

Because Lowell is a less densley populated area than Boston, I think it would be reasonable to say that there should be no gas stations within 150 meters of a school.

```python
minDist=150
```

```python
# create buffer around each school area
schoolsLowell_buffered=schoolsLowell.buffer(minDist)
schoolsLowell_buffered
```

```python
# turn geoseries into geodataframe
secured_schoolsLowell=gpd.GeoDataFrame(geometry=schoolsLowell_buffered)
secured_schoolsLowell
```

```python
# number of polygons created with buffers
secured_schoolsLowell.geometry.geom_type.value_counts()
```

```python
# gas stations in Lowell - points or polygons?
fuelLowell.geometry.geom_type.value_counts()
```

```python
# create buffer around each gas stations to ensure proper selection
fuelLowell_allPoly=fuelLowell.copy()
fuelLowell_allPoly['geometry'] = [station.buffer(10) if station.geom_type == 'Point' else station
                                  for station in fuelLowell.geometry]
```

```python
fuelLowell_allPoly.geometry.geom_type.value_counts()

# now all stations are polygons
```

```python
# sjoin to determine gas stations that need to move
gas_relocate=fuelLowell_allPoly.sjoin(secured_schoolsLowell,how='inner',predicate='intersects')
```

```python
# check for duplicates
is_duplicate=gas_relocate.index.duplicated(keep=False)
gas_relocate[is_duplicate]

# no duplicates
```

```python
# school buffers and stations to relocate
base=secured_schoolsLowell.sjoin(gas_relocate,how='inner',predicate='intersects').explore(color='yellow', tiles='cartodbpositron')
gas_relocate.explore(m=base,color='red')
```

```python
# find new locations for gas stations

place = "Lowell, Massachusetts, USA"

G = ox.graph_from_place(place, network_type='drive')

streets = ox.graph_to_gdfs(G, nodes=False, edges=True)

streets.shape
```

```python
streets.info()
```

```python
# focus on highways
streets.highway.value_counts().index
```

```python
GAS_STATION_ROAD_TYPES = [
    'motorway',      # High-speed limited access roads
    'primary',       # Major national roads
    'trunk',         # Important regional roads
    'motorway_link', # Motorway entrance/exit ramps
    'primary_link',  # Primary road connectors
    'trunk_link',     # Trunk road connectors
    ['motorway', 'trunk'],
    ['primary', 'motorway_link'],
    ['primary', 'primary_link']
]

suitable_roads = streets[streets.highway.isin(GAS_STATION_ROAD_TYPES)]

suitable_roads.shape
```

```python
# reproject suitable roads to crs
suitable_roads = suitable_roads.to_crs(lowell_crs)
```

```python
# new locations must be far from schools and other gas stations
# create buffer around gas stations
compDistance=150
fuelLowell_allPoly_buffered=fuelLowell_allPoly.buffer(compDistance)
```

```python
type(fuelLowell_allPoly_buffered)
secured_fuelLowell_allPoly=gpd.GeoDataFrame(geometry=fuelLowell_allPoly_buffered)
```

```python
# combine secured areas
secured_areas_dissolved=secured_fuelLowell_allPoly.overlay(secured_schoolsLowell,how='union',keep_geom_type=False).dissolve()
secured_areas=gpd.GeoDataFrame(geometry=secured_areas_dissolved.geometry)
```

```python
# filter roads using overlay
suitable_roads.overlay(secured_areas,keep_geom_type=False, how='difference').plot()
```

```python
good_routes_forGas_dissolved=suitable_roads.overlay(secured_areas,keep_geom_type=False, how='difference').dissolve()
good_routes_forGas=gpd.GeoDataFrame(geometry=good_routes_forGas_dissolved.geometry)
good_routes_forGas

good_routes_forGas.plot()
```

```python
# now we can choose a location from these places
potential_Locations=good_routes_forGas.copy()
newPOLYGONS=[]
pointsComputed=1
attempts=1
while pointsComputed<=len(gas_relocate):
  candidatePoint=potential_Locations.sample_points(1)
  candidateStation=gpd.GeoDataFrame(geometry=candidatePoint.buffer(100))
  if candidateStation.overlay(secured_areas,how='intersection',keep_geom_type=False).empty:
    newPOLYGONS.append(candidateStation)
    pointsComputed+=1
    potential_Locations=potential_Locations.overlay(candidateStation, how='difference', keep_geom_type=False)
  if attempts>len(gas_relocate)*10:
    break
  attempts+=1

attempts,len(newPOLYGONS)
```

```python
newPOLYGONS
```

```python
import pandas as pd
pd.concat(newPOLYGONS)
```

```python
GoodLocationsGas_buffered_gdf=pd.concat(newPOLYGONS)
GoodLocationsGas_gdf=gpd.GeoDataFrame(geometry=GoodLocationsGas_buffered_gdf.centroid)
```

```python
# original suitable roads
base = suitable_roads.plot(color='yellow', figsize=(10, 10),zorder=1)

# non-overlapping buffers
GoodLocationsGas_buffered_gdf.plot(ax=base, edgecolor='red', marker="+", facecolor='none',zorder=2)

# center points of the buffers
GoodLocationsGas_gdf.plot(ax=base, color='red', marker="+", markersize=50,zorder=3)

# all stations
fuelLowell.plot(ax=base,color='k',zorder=4)
schoolsLowell.plot(ax=base,color='green',zorder=5)

# stations to relocate
fuelLowell.clip(gas_relocate).plot(ax=base,color='magenta',zorder=5)
```

Reverse geocoding

```python
# get addresses
from geopy.geocoders import Nominatim
import time

# Initialize geocoder
geolocator = Nominatim(user_agent="lowell_gas_stations")

def get_address(point):
  # Now coordinates are in degrees (lon, lat)
  lon = point.x
  lat = point.y
  location = geolocator.reverse((lat, lon), exactly_one=True, timeout=10)
  time.sleep(1)  # Rate limiting

  if location:
    return location.address
  else:
    return None # not found

GoodLocationsGas_4326 = GoodLocationsGas_gdf.to_crs('EPSG:4326')
GoodLocationsGas_gdf['address'] = GoodLocationsGas_4326.geometry.apply(get_address)

GoodLocationsGas_gdf
```

```python
suitable_roads.to_file('lowell_analysis.gpkg', layer='suitable_roads', driver='GPKG')
GoodLocationsGas_buffered_gdf.to_file('lowell_analysis.gpkg', layer='gas_buffered', driver='GPKG')
GoodLocationsGas_gdf.to_file('lowell_analysis.gpkg', layer='gas_centers', driver='GPKG')
fuelLowell.to_file('lowell_analysis.gpkg', layer='fuel_stations', driver='GPKG')
schoolsLowell.to_file('lowell_analysis.gpkg', layer='schools', driver='GPKG')
fuelLowell.clip(gas_relocate).to_file('lowell_analysis.gpkg', layer='fuel_relocate', driver='GPKG')
lowell.to_file('lowell_analysis.gpkg', layer='lowell', driver='GPKG')
```
